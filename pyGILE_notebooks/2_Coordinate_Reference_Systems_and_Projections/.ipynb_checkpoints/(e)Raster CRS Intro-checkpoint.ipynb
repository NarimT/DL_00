# Raster Coordinate Reference Systems (CRS)

Raster data is very different that vector data, one of the key differences is that we don't have a pair of coordinates (x,y) for each pixel in a raster. How then do we know where the raster is located in addition to what the data values are? For a new spatial raster (e.g. geotif) we need to store a few other pieces of information seperately. We need to keep track of the location of the upper left hand corner, the resolution (in both the x and y direction) and a description of the coordinate space (i.e. the CRS), amongst others.

Let's start from the `ndarray` `Z` that we want to span from [-90&deg;,90&deg;] longitude, and [-90&deg;,90&deg;] latitude.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-90, 90, 6)
y = np.linspace(90, -90, 6)
X, Y = np.meshgrid(x, y)

Z1 =  np.abs(((X - 10) ** 2 + (Y - 10) ** 2) / 1 ** 2)
Z2 =  np.abs(((X + 10) ** 2 + (Y + 10) ** 2) / 2.5 ** 2)
Z =  (Z1 - Z2)

plt.imshow(Z)
plt.title("Temperature")
plt.show()
```

Note that *`Z` contains no data on its location*. Its just an array, the information stored in `x` and `y` aren't associated with it at all. This location data will often be stored in the header of file. In order to 'locate' the array on the map we will use affine transformations. 

Affine transformations allows us to use simple systems of linear equations to manipulate any point or set of points [review affine transforms]. It allows us to move, stretch, or even rotate a point or set of points. In the case of GIS, it is used to move raster data, a satellite image, to the correct location in the CRS coordinate space.

## Describing the Array Location (Define a Projection)
In this example the coordinate reference system will be '+proj=latlong', which describes an equirectangular coordinate reference system with units of decimal degrees. Although `X` and `Y` seems relevant to understanding the location of cell values, `rasterio` instead uses affine transformations instead. Affine transforms uses matrix algebra to describe where a cell is located (translation) and what its resolution is (scale).

The affine transformation matrix can be computed from the matrix product of a translation (moving N,S,E,W) and a scaling (resolution). First, we start with translation where $\Delta x$ and $\Delta y$ define the location of the upper left hand corner of our new `Z` ndarray. As a reminder the translation matrix takes the form:

$$
   \begin{eqnarray}
     \mbox{Translate} =  \begin{bmatrix} 1 & 0 & \Delta x \\  0 & 1 & \Delta y \\ 0 & 0 & 1 \end{bmatrix}  
   \end{eqnarray}
$$

Now we can define our translation matrix using the point coordinates (`x[0]`,`y[0]`), but these need to be offset by 1/2 the resolution so that the cell is centered over the coordinate (-90,90). Notice however there are some difference between the x and y resolution:

Both arrays have the same spatial resolution

```python
xres = (x[-1] - x[0]) / len(x)
xres
```

But notice that the y resolution is **negative**:

```python
yres = (y[-1] - y[0]) / len(y)
yres
```

We need to create our translation matrix by defining the location of the upper left hand corner. Looking back at our definitions of our coordinates `X` and `Y` we can see that they are defined with `x = np.linspace(-90, 90, 6); y = np.linspace(90, -90, 6); X, Y = np.meshgrid(x, y)`, if you run `print(X);print(Y)` you will see that the **center** of the upper left hand corner should be located at (-90,90). The **upper left hand corner** of that same cell is actually further up and to the left than (-90,90). It follows then that that corner should be shifted exactly 1/2 the resolution of the cell, both up and to the left. 

We can therefore define the upper left hand corner by setting $\Delta x = x[0] - xres / 2$ and $\Delta y = y[0] - yres / 2$. Remember yres is negative, so subtraction is correct.

```python
from rasterio.transform import Affine
print(Affine.translation(x[0] - xres / 2, y[0] - yres / 2))
```

We also need to scale our data based on the resolution of each cell, the scale matrix takes the following form:

$$
   \begin{eqnarray}
   \mbox{Scale} = \begin{bmatrix} xres & 0 & 0 \\ 0 & yres & 0 \\ 0 & 0 & 1 \end{bmatrix} 
   \end{eqnarray}
$$

```python
print(Affine.scale(xres, yres))
```

We can do both operations simultaneously in a new `transform` matrix by calculating the product of the Translate and Scale matrix: 

$$
    \mbox{Translate} \cdot \mbox{Scale} =  \begin{bmatrix} xres & 0 & \Delta x \\ 0 & yres & \Delta y \\ 0 & 0 & 1 \end{bmatrix} 
$$

```python
transform = Affine.translation(x[0] - xres / 2, y[0] - yres / 2) * Affine.scale(xres, yres)
print(transform)
```

Now we need to write out a `tif` file that holds the data in `Z` and its data type with `dtype`, the location described by `transform`, in coordinates described by the coordinate reference system `+proj=latlon`, the number of 'bands' of data in `count` (in this case just one), and the shape in `height` and `width`.

```python
import rasterio
with rasterio.open(
    '../../pygis/temp/Z.tif',
    'w',
    driver='GTiff',
    height=Z.shape[0],
    width=Z.shape[1],
    count=1,
    dtype=Z.dtype,
    crs='+proj=latlong',
    transform=transform,
) as dst:
    dst.write(Z, 1)
```

## Writing Georeferenced Raster Data

All this geographic information is stored in the `dst` object and then written to disk with `dst.write(Z, 1)`. The `write` method takes the array of data `Z` and the band number to write to (in this case band `1`). This syntax may seem awkward, but it's a carryover from GDAL, which rasterio relies on heavily (like most other GIS platforms including ArcGIS).

## Understanding Raster Positioning with Transforms

When we save our temperature array `Z` to a GeoTIFF file, we need to tell the file where each pixel belongs on Earth's surface. The `transform` parameter accomplishes this using an affine transformation matrix that converts array indices (row, column) to geographic coordinates (x, y).

### The Transform Matrix

Our transform matrix acts as a "map" that converts pixel positions to real-world coordinates. For our example:
- Resolution: 30 degrees per pixel
- Upper left corner location: (-105, 105)
- Transform matrix: `[30, 0, -105, 0, -30, 105]`

This represents the 3x3 affine transformation matrix:
$$\begin{bmatrix} 30 & 0 & -105 \\ 0 & -30 & 105 \\ 0 & 0 & 1 \end{bmatrix}$$

### Why the Upper Left Corner is at (-105, 105)

To understand the transform, we need to work through positioning our temperature raster data `Z`. Our goal is to center the upper left cell over the coordinate (-90, 90). However, the transform matrix references the upper left corner of that cell, not its center.

Since each cell has a resolution of 30 degrees, the upper left corner needs to be offset by half the resolution (15 degrees) both west and north of the cell center:
- West offset: -90 - 15 = -105
- North offset: 90 + 15 = 105

The calculation gives us the final coordinates of the upper left corner: $(x_0, y_0) = (-105, 105)$

### Transform as a Coordinate Mapping System

The transform matrix allows us to find the coordinates of any cell based on its row and column numbers. We multiply the transform matrix by the column and row indices to retrieve the geographic coordinates of that cell's upper left corner.

The formula is: **Geographic Coordinates = Transform Matrix × [Column, Row, 1]**

### Working Through Coordinate Calculations

Let's calculate the upper left corner coordinates for specific cells using our transform matrix. Remember, the matrix multiplication formula is:

$\text{Geographic Coordinates} = \begin{bmatrix} x_{res} & 0 & x_{offset} \\ 0 & y_{res} & y_{offset} \\ 0 & 0 & 1 \end{bmatrix} \begin{bmatrix} \text{column} \\ \text{row} \\ 1 \end{bmatrix}$

**Example 1: Upper left cell (row=0, column=0)**

**Step 1:** Set up the multiplication
$\begin{bmatrix} 30 & 0 & -105 \\ 0 & -30 & 105 \\ 0 & 0 & 1 \end{bmatrix} \begin{bmatrix} 0 \\ 0 \\ 1 \end{bmatrix}$

**Step 2:** Calculate each row using matrix multiplication rules
- **Row 1 (x-coordinate):** $(30 \times 0) + (0 \times 0) + (-105 \times 1) = 0 + 0 + (-105) = -105$
- **Row 2 (y-coordinate):** $(0 \times 0) + (-30 \times 0) + (105 \times 1) = 0 + 0 + 105 = 105$
- **Row 3 (homogeneous):** $(0 \times 0) + (0 \times 0) + (1 \times 1) = 1$

**Step 3:** Write the result
$\begin{bmatrix} -105 \\ 105 \\ 1 \end{bmatrix}$

**Result:** Upper left corner of cell (0,0) is at geographic coordinates (-105, 105)

**Example 2: Cell at row=1, column=2**

**Step 1:** Set up the multiplication
$\begin{bmatrix} 30 & 0 & -105 \\ 0 & -30 & 105 \\ 0 & 0 & 1 \end{bmatrix} \begin{bmatrix} 2 \\ 1 \\ 1 \end{bmatrix}$

**Step 2:** Calculate each row step by step
- **Row 1 (x-coordinate):** 
  - $(30 \times 2) + (0 \times 1) + (-105 \times 1)$
  - $= 60 + 0 + (-105)$
  - $= 60 - 105 = -45$

- **Row 2 (y-coordinate):**
  - $(0 \times 2) + (-30 \times 1) + (105 \times 1)$
  - $= 0 + (-30) + 105$
  - $= -30 + 105 = 75$

- **Row 3 (homogeneous):**
  - $(0 \times 2) + (0 \times 1) + (1 \times 1) = 1$

**Step 3:** Write the result
$\begin{bmatrix} -45 \\ 75 \\ 1 \end{bmatrix}$

**Result:** Upper left corner of cell (1,2) is at geographic coordinates (-45, 75)

**Example 3: Let's verify with cell at row=2, column=3**

**Step 1:** Set up the multiplication
$\begin{bmatrix} 30 & 0 & -105 \\ 0 & -30 & 105 \\ 0 & 0 & 1 \end{bmatrix} \begin{bmatrix} 3 \\ 2 \\ 1 \end{bmatrix}$

**Step 2:** Calculate each row
- **Row 1 (x-coordinate):** $(30 \times 3) + (0 \times 2) + (-105 \times 1) = 90 + 0 - 105 = -15$
- **Row 2 (y-coordinate):** $(0 \times 3) + (-30 \times 2) + (105 \times 1) = 0 - 60 + 105 = 45$
- **Row 3 (homogeneous):** $(0 \times 3) + (0 \times 2) + (1 \times 1) = 1$

**Result:** Upper left corner of cell (2,3) is at geographic coordinates (-15, 45)

### Understanding the Pattern

Notice the pattern in our results:
- Moving right (increasing column): x-coordinate increases by 30 degrees per column
- Moving down (increasing row): y-coordinate decreases by 30 degrees per row

This systematic approach allows any GIS software to precisely locate every pixel in your raster on Earth's surface, enabling accurate analysis and display of your temperature data. The transform matrix essentially creates a mathematical "address system" for every pixel in your raster array.

```python
print(transform*(0,0))
```

Just to make sure it works let's find a harder one, 5th column right, 2nd row down:

#### How Transform Matrix Works

Let's work through an example of finding the upper left coordinates for `row=2`, `column=5`:

**Step 1:** Set up the matrix multiplication
$$\begin{bmatrix} 
    30 & 0 & -105 \\  
    0 & -30 & 105 \\ 
    0 & 0 & 1 
\end{bmatrix}  
\begin{bmatrix} 5 \\ 2 \\ 1 \end{bmatrix}$$

**Step 2:** Calculate each row
- **Row 1 (x-coordinate):** $(5 \times 30) + (2 \times 0) + (1 \times -105) = 150 + 0 - 105 = 45$
- **Row 2 (y-coordinate):** $(5 \times 0) + (2 \times -30) + (1 \times 105) = 0 - 60 + 105 = 45$  
- **Row 3 (homogeneous):** $(5 \times 0) + (2 \times 0) + (1 \times 1) = 1$

**Result:** 
$$\begin{bmatrix} 45 \\ 45 \\ 1 \end{bmatrix}$$

The upper left corner of the cell at row=2, column=5 is located at geographic coordinates (45, 45).

## Raster Reprojection Fundamentals

How do we reproject a raster from one coordinate system to another? Since the `transform` matrix maps pixel locations to geographic coordinates, raster reprojection involves calculating a new `transform` for the destination coordinate reference system (CRS) and then mapping pixels accordingly.

### Simple Case: Shifting the Prime Meridian

One straightforward reprojection case involves false easting or shifting the prime meridian. Let's examine an example where we:
- Start with a raster having its upper left corner at (0, 45)
- Reproject it by moving the prime meridian 10° west (using `+lon_0=-10` in proj4)
- Result in the upper left corner moving to (10, 45)

### Complex Case: Geometric Transformation and Warping

In many reprojection scenarios, we must change the number of rows or columns, or 'warp' the image geometry. This creates a fundamental challenge: the pixel centers of the new projected raster don't align with the pixel centers of the original raster. Consequently, the new pixels represent ground locations that weren't directly sampled in the original dataset.

Consider switching from a geographic CRS (latitude/longitude) to a projected CRS (flat coordinate system). The pixel centers no longer overlap between the two coordinate systems:

<div align="center">

<br><em>Misaligned pixel centers after coordinate transformation</em>
</div>

This misalignment requires us to estimate values for the new pixel locations using interpolation methods.

### Interpolation Methods

When pixel centers don't align, we need interpolation to estimate values at new locations. Three common methods are available:

**1. Nearest Neighbor:** Assigns the value of the closest original pixel

**2. Bilinear:** Uses linear interpolation between nearby pixels  

**3. Bicubic:** Uses curved interpolation for smoother transitions

Here's how these methods estimate a value for location `C` based on known values at locations `A` and `B`:

### Choosing the Appropriate Interpolation Method

The choice of interpolation method depends on your data type and quality requirements:

| Method | Description | Speed | Categorical Data | Continuous Data |
|--------|-------------|-------|------------------|-----------------|
| Nearest Neighbor | Assigns nearest pixel value | Fast | ✓ Recommended | Limited use |
| Bilinear | Linear interpolation between pixels | Fast | ✗ Not suitable | ✓ Good choice |
| Bicubic | Curved interpolation | Moderate | ✗ Not suitable | ✓ Best quality |

**For categorical data** (land cover classes, administrative boundaries): Use Nearest Neighbor exclusively. Other methods will create invalid intermediate values.

**For continuous data** (temperature, elevation, rainfall): Choose between Bilinear and Bicubic. Bilinear provides good results with fast processing. Bicubic offers superior quality for data with complex spatial patterns, though it requires more computation time.

## Practical Reprojection with Rasterio

To implement raster reprojection, we use rasterio's `calculate_default_transform` function. This function generates the transform matrix required for the new coordinate system based on:
- Source raster characteristics
- Destination CRS specifications
- Desired output resolution and extent

The transform matrix follows this standard form:

$$\text{Transform} = \begin{bmatrix} x_{res} & 0 & \Delta x \\ 0 & y_{res} & \Delta y \\ 0 & 0 & 1 \end{bmatrix}$$

Where:
- $x_{res}, y_{res}$ = pixel resolution in x and y directions
- $\Delta x, \Delta y$ = upper left corner coordinates

Let's examine how to calculate the transform matrix for a destination raster in the next section.

```python
import numpy as np
import rasterio
from rasterio.warp import reproject, Resampling, calculate_default_transform

dst_crs = "EPSG:3857"  # web mercator(ie google maps)

with rasterio.open("../../pygis/data/LC08_L1TP_224078_20200518_20200518_01_RT.TIF") as src:

    # transform for input raster
    src_transform = src.transform

    # calculate the transform matrix for the output
    dst_transform, width, height = calculate_default_transform(
        src.crs,    # source CRS
        dst_crs,    # destination CRS
        src.width,    # column count
        src.height,  # row count
        *src.bounds,  # unpacks outer boundaries (left, bottom, right, top)
    )

print("Source Transform:\n",src_transform,'\n')
print("Destination Transform:\n", dst_transform)
```

Notice that in order to keep the same number of rows and columns that the resolution of the destination raster increased from 30 meters to 33.24 meters. Also the coordinates of the upper left hand corner have shifted (check $\Delta x, \Delta y$).

Ok finally!

```python
dst_crs = "EPSG:3857"  # web mercator(ie google maps)

with rasterio.open("../../pygis/data/LC08_L1TP_224078_20200518_20200518_01_RT.TIF") as src:
    src_transform = src.transform

    # calculate the transform matrix for the output
    dst_transform, width, height = calculate_default_transform(
        src.crs,
        dst_crs,
        src.width,
        src.height,
        *src.bounds,  # unpacks outer boundaries (left, bottom, right, top)
    )

    # set properties for output
    dst_kwargs = src.meta.copy()
    dst_kwargs.update(
        {
            "crs": dst_crs,
            "transform": dst_transform,
            "width": width,
            "height": height,
            "nodata": 0,  # replace 0 with np.nan
        }
    )

    with rasterio.open("../../pygis/temp/LC08_20200518_webMC.tif", "w", **dst_kwargs) as dst:
        for i in range(1, src.count + 1):
            reproject(
                source=rasterio.band(src, i),
                destination=rasterio.band(dst, i),
                src_transform=src.transform,
                src_crs=src.crs,
                dst_transform=dst_transform,
                dst_crs=dst_crs,
                resampling=Resampling.nearest,
            )
```
